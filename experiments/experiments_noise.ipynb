```python
from runner import main

logs = {
    "bpi2012": {
        "proxyLog": "logs/bpi2012/bpi2012_100traces.xes.gz",
        "numOfEvents": "1966",
        "timeoutInSec": "70",
        },
    "bpi2017": {
        "proxyLog": "logs/bpi2017/bpi2017_100traces.xes.gz",
        "numOfEvents": "3995",
        "timeoutInSec": "500",
        },
    "bpi2020_travelpermits": {
        "proxyLog": "logs/bpi2020_travelpermits/bpi2020_travelpermits_100traces.xes.gz",
        "numOfEvents": "1205",
        "timeoutInSec": "40",
        },
}

algo_type = {
    "ngrams": {
        "backToTheOrder": "ngrams",
        "eventTimeAware": "true",
        "adaptable": "true",
    },
    "greedy": {
        "backToTheOrder": "greedy",
        "eventTimeAware": "true",
        "adaptable": "true",
    },
    "adaptive": {
        "backToTheOrder": "none",
        "eventTimeAware": "true",
        "adaptable": "true",
    },
    "non_adaptive": {
        "backToTheOrder": "none",
        "eventTimeAware": "true",
        "adaptable": "false",
    },
    "non_aware": {
        "backToTheOrder": "none",
        "eventTimeAware": "false",
        "adaptable": "false",
    },
}

swap_options = ["01", "10", "50"]
noise_options = ["05","10","20"]
```

```python
def call_runner(proxyLog,dataset,algo_type,backToTheOrder,eventTimeAware,adaptable,swap,numOfEvents,timeoutInSec):
    custom_args = [
                    "--proxyLog", proxyLog,
                    "--logFriendlyName", f"{dataset}__algo_{algo_type}__swap_{swap}",
                    "--minDecayTime", "3",
                    "--decayTimeMultiplier", "0.3",
                    "--ewmaAlpha", "0.005",
                    "--eventTimeAware", eventTimeAware,
                    "--adaptable", adaptable,
                    "--maxPatternLength", "10",
                    "--backToTheOrder", backToTheOrder,
                    "--cores", "1",
                    "--swap_setting", swap,
                    "--dataset", dataset,
                    "--numOfEvents", numOfEvents,
                    "--timeoutInSec", timeoutInSec
                ]

    main(custom_args)
```

```python
for k,v in logs.items():    
    for _noise in noise_options:   
        _dataset = f"{k}__noise_{_noise}"      
        for _algo_k, _algo_v in algo_type.items():
            if _algo_k not in ["non_adaptive","non_aware"]:    #tmp
                continue                #tmp
            for _swap in swap_options:
                for _i in range(3):
                    call_runner(v["proxyLog"], _dataset, _algo_k, _algo_v["backToTheOrder"], _algo_v["eventTimeAware"], _algo_v["adaptable"], _swap, v["numOfEvents"], v["timeoutInSec"])
```
