```python
from runner import main

logs = {
    "bpi2012": {
        "proxyLog": "logs/bpi2012/bpi2012_100traces.xes.gz",
        "numOfEvents": "1966",
        "timeoutInSec": "60",
        },
    "bpi2017": {
        "proxyLog": "logs/bpi2017/bpi2017_100traces.xes.gz",
        "numOfEvents": "3995",
        "timeoutInSec": "600",
        },
    "bpi2020_travelpermits": {
        "proxyLog": "logs/bpi2020_travelpermits/bpi2020_travelpermits_100traces.xes.gz",
        "numOfEvents": "1205",
        "timeoutInSec": "30",
        },
}

ewma_options = ["0.0005","0.001","0.0025","0.005","0.01","0.025","0.05"]

swap_options = ["01","10","50"]
```

```python
def call_runner(proxyLog,dataset,ewma,swap,numOfEvents,timeoutInSec):
    custom_args = [
                    "--proxyLog", proxyLog,
                    "--logFriendlyName", f"{dataset}__ewma_{ewma}__swap_{swap}",
                    "--minDecayTime", "3",
                    "--decayTimeMultiplier", "0.3",
                    "--ewmaAlpha", ewma,
                    "--eventTimeAware", "true",
                    "--adaptable", "true",
                    "--maxPatternLength", "10",
                    "--backToTheOrder", "ngrams",
                    "--cores", "1",
                    "--swap_setting", swap,
                    "--dataset", dataset,
                    "--numOfEvents", numOfEvents,
                    "--timeoutInSec", timeoutInSec
                ]

    main(custom_args)
```

```python
for k,v in logs.items():
    for _ewma in ewma_options:
        for _swap in swap_options:
            
            for _i in range(3):
                call_runner(v["proxyLog"], k, _ewma, _swap, v["numOfEvents"], v["timeoutInSec"])
```
